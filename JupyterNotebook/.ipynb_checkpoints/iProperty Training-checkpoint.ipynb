```python
import pandas as pd
import numpy as np

from sklearn.linear_model import LinearRegression

from sklearn.model_selection import train_test_split

from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler

from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_absolute_percentage_error
```

```python
df = pd.read_csv('Cleaned.csv')
```

```python
df
```

```python
one_hot = pd.get_dummies(df[['type']])

one_hot
```

```python
df.drop(['type'],axis=1,inplace=True)
df = df.join(one_hot)
```

```python
def remove_outliers(dataframe, col_name):
    for col in col_name:
        Q1 = dataframe[col].quantile(0.25)
        Q3 = dataframe[col].quantile(0.75)
        IQR = Q3 - Q1      
        filtered = dataframe.query(f'(@Q1 - 1.5 * @IQR) <= {col} <= (@Q3 + 1.5 * @IQR)')
    return filtered
```

```python
df = remove_outliers(df, ['price','area'])
```

```python
df.describe()
```

```python
df.shape
```

```python
X = df.values[:,1:]
y = df.values[:,0]
```

```python
X
```

```python
y
```

# Standard scaling the X

```python
std1 = StandardScaler()
X_std = std1.fit_transform(X)
```

# Train Test Split

```python
X_train_std , X_test_std , y_train, y_test = train_test_split(X_std, y, test_size=0.2, random_state=42)
```

# Training the model

```python
reg = LinearRegression()
```

```python
reg.fit(X_train_std, y_train)
```

```python
y_pred = reg.predict(X_test_std)
```

# Mean_Squared_Error

```python
mean_squared_error(y_test, y_pred)
```

# Root_Mean_Squared_Error

```python
mean_squared_error(y_test, y_pred, squared=False)
```

# Mean_Absolute_Error

```python
mean_absolute_error(y_test, y_pred)
```

# Mean_Absolute_Percentage_Error

```python
mean_absolute_percentage_error(y_test, y_pred)
```
